```python
df_cars_2019 = pd.read_excel("data-raw/vmrc_anual_tr_cifra_2019.xlsx")
df_cars_2019
```

```python

df_cars_2019['CVE_ENT'] = df_cars_2019['ID_ENTIDAD'].astype(str).str.zfill(2)

df_cars_2019['CVE_MUN'] = df_cars_2019['ID_MUNICIPIO'].astype(str).str.zfill(3)

df_cars_2019['CVEGEO'] = df_cars_2019['CVE_ENT'] + df_cars_2019['CVE_MUN']

df_cars_2019['vehicles'] = df_cars_2019['AUTO_OFICIAL'] + df_cars_2019['AUTO_PUBLICO'] + df_cars_2019['AUTO_PARTICULAR'] + df_cars_2019['CAM_PAS_OFICIAL'] + df_cars_2019['CAM_PAS_PUBLICO'] + df_cars_2019['CAM_PAS_PARTICULAR'] + df_cars_2019['CYC_CARGA_OFICIAL'] + df_cars_2019['CYC_CARGA_PUBLICO'] + df_cars_2019['CYC_CARGA_PARTICULAR'] + df_cars_2019['MOTO_OFICIAL'] + df_cars_2019['MOTO_DE_ALQUILER'] + df_cars_2019['MOTO_PARTICULAR'] 



df_cars_2019_f = df_cars_2019[["CVEGEO", "vehicles"]]

df_cars_2019_f
```

```python
df_cars_2019_f.to_excel('data/df_cars_2019.xlsx', index = False)
df_cars_2019_f.to_parquet('data/df_cars_2019.parquet')
```

```python
dict_SEDENA = {"MILITAR_MUERTO_NUM": "deaths_militar","MILITAR_HERIDO_NUM":"injur_militar",
                       "AGRESOR_MUERTO_NUM":"deaths_aggressor", "AGRESOR_HERIDO_NUM":"injur_aggressor",
               "AGRESOR_DETENIDO_NUM":"arrest_aggressor", "CLAVE_MUNICIPIO": "CVEGEO", "enfren": "confront"} 

df_SEDENA_08_19_w = df_SEDENA_08_19[["year", "CLAVE_MUNICIPIO", "MILITAR_HERIDO_NUM", "MILITAR_MUERTO_NUM", "AGRESOR_MUERTO_NUM", "AGRESOR_HERIDO_NUM", "AGRESOR_DETENIDO_NUM", "enfren"]].rename(columns = dict_SEDENA)
```

```python
df_SEDENA_08_19_w
```

```python
df_SEDENA_08_19_w_group = df_SEDENA_08_19_w.groupby(['year', 'CVEGEO'], as_index=0).sum()
```

```python
df_SEDENA_08_19_w_group
```

```python
df_SEDENA_group_15_19 = df_SEDENA_08_19_w_group[(df_SEDENA_08_19_w_group['year']>=2015)]
```

```python
df_SEDENA_group_15_19
```

```python
df_SEDENA_group_15_19["mortality_index"] = df_SEDENA_group_15_19["deaths_aggressor"]/df_SEDENA_group_15_19["injur_aggressor"]
```

```python
df_SEDENA_group_15_19_w = df_SEDENA_group_15_19.groupby(['CVEGEO'], as_index=0).mean()
```

```python
df_SEDENA_group_15_19_w_f = df_SEDENA_group_15_19_w[["CVEGEO", "confront"]]
```

```python
df_SEDENA_group_15_19_w_f["confront"] = df_SEDENA_group_15_19_w["confront"].round()
```

```python
df_SEDENA_group_15_19_w_f
```
