```python
df_cars_2015 = pd.read_excel("data-raw/vmrc_anual_tr_cifra_2015.xlsx")
df_cars_2016 = pd.read_excel("data-raw/vmrc_anual_tr_cifra_2016.xlsx")
df_cars_2017 = pd.read_excel("data-raw/vmrc_anual_tr_cifra_2017.xlsx")
df_cars_2018 = pd.read_excel("data-raw/vmrc_anual_tr_cifra_2018.xlsx")
df_cars_2019 = pd.read_excel("data-raw/vmrc_anual_tr_cifra_2019.xlsx")



```

```python
def add_variables(df):
    
    df['CVE_ENT'] = df['ID_ENTIDAD'].astype(str).str.zfill(2)
    
    df['CVE_MUN'] = df['ID_MUNICIPIO'].astype(str).str.zfill(3)
    
    df['CVEGEO'] = df['CVE_ENT'] + df['CVE_MUN']
    
    df['vehicles'] = df['AUTO_OFICIAL'] + df['AUTO_PUBLICO'] + df['AUTO_PARTICULAR'] + df['CAM_PAS_OFICIAL'] + df['CAM_PAS_PUBLICO'] + df['CAM_PAS_PARTICULAR'] + df['CYC_CARGA_OFICIAL'] + df_cars_2019['CYC_CARGA_PUBLICO'] + df_cars_2019['CYC_CARGA_PARTICULAR'] + df_cars_2019['MOTO_OFICIAL'] + df_cars_2019['MOTO_DE_ALQUILER'] + df_cars_2019['MOTO_PARTICULAR'] 
    
    df_f = df[["CVEGEO", "vehicles"]]
    
    return df_f


df_list = [df_cars_2015, df_cars_2016, df_cars_2017, df_cars_2018, df_cars_2019]

df_cars_2015, df_cars_2016, df_cars_2017, df_cars_2018, df_cars_2019 = [add_variables(df) for df in df_list]


```

```python


dataframes = [df_cars_2015, df_cars_2016, df_cars_2017, df_cars_2018, df_cars_2019]




df_vehicles_final = reduce(lambda  left,right: pd.merge(left,right,on=['CVEGEO'],
                                            how = 'outer'), dataframes)#.fillna(0)

df_vehicles_final = df_vehicles_final.fillna(0)

df_vehicles_final
```

```python
dict_vec = {"S_2015": "vehicles_x"} 

df_vehicles_final_w = df_vehicles_final.rename(columns = dict_vec)

df_vehicles_final_w.iloc[:, 1] + df_vehicles_final_w.iloc[:, 2]
```

```python
df_vehicles_final["vehicles_15_19"] = df_vehicles_final_w.iloc[:, 1] + df_vehicles_final_w.iloc[:, 2] + df_vehicles_final_w.iloc[:, 3] + df_vehicles_final_w.iloc[:, 4] + df_vehicles_final_w.iloc[:, 5]/5

df_vehicles_final = df_vehicles_final[["CVEGEO", "vehicles_15_19"]]

df_vehicles_final
```
