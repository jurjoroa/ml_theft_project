```python
df_social_2015_ori = pd.read_excel("data-raw/Mexico - Social Indicators - IMM_2015.xlsx", converters={'CVE_ENT' : str, 'CVE_MUN' : str})
df_social_2020_ori = pd.read_excel("data-raw/Mexico - Social Indicators - IMM_2020.xlsx", converters={'CVE_ENT' : str, 'CVE_MUN' : str})
```
