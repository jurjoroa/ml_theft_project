```python
#Upload the CSV file
import pandas as pd


from google.colab import files

uploaded = files.upload()
```

```python
#Create a dataframe with the csv uploaded
import io
df_original = pd.read_csv(io.BytesIO(uploaded['df_thefts_mun_code.csv']))
```

```python
#Check the first 5 rows
df_original.head()
```

```python
#create a new dataframe and rename the variables
df_pipethefts = df_original.rename(columns = {'year' : 'Year', 'state' : 'State', 'municipio' : 'Municipality', 'tomas' : 'Num_thefts', 'codigo inegi': 'Mun_code'})

#reorder the variables
df_pipethefts = df_pipethefts.iloc[:, [4,1,2,0,3]]
df_pipethefts
```

```python
#Delete the rows with code 'ERROR'(not valid unique CODE)
df_pipethefts.drop(df_pipethefts[df_pipethefts['Mun_code'] == 'ERROR'].index, axis = 0, inplace = True)

# Delete the variables 'State' and 'Municipality' (We don't need that as we have the Muninicipality code)
df_pipethefts.drop(['State', 'Municipality'], axis = 1, inplace = True)
```

```python

# Basics informations about the dataframe
print(df_pipethefts)
print(sum(df_pipethefts['Num_thefts']))
print(len(df_pipethefts))
print(df_pipethefts.describe())
```

```python
#Find duplicated rows using Mun_code and year

duplicateDFRow = df_pipethefts[df_pipethefts.duplicated(['Mun_code','Year'], keep = False)]
duplicateDFRow
```

```python
#group the database
df_pipethefts_grouped = df_pipethefts.groupby(['Mun_code','Year'], as_index=False, sort=True)['Num_thefts'].sum()
df_pipethefts_grouped
```

```python
# Year filtering and mean calculation

df_pipethefts_years_mean = df_pipethefts[df_pipethefts['Year'] >= 2010].groupby('Mun_code', as_index=False, sort=True)['Num_thefts'].mean()
df_pipethefts_years_mean

#Export df_pipethefts_years_mean to a csv file
#df_pipethefts_years_mean.to_csv
```
